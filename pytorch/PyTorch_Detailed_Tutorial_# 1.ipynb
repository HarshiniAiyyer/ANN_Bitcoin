#### Reshaping, stacking, Squeezing and unsqueezing tensors

- reshaping : fixing shapes / dimensions
- view -> return a view of the input tnesor but the same memory as the original tensor
- stacking -> multi tensors on top of the other (vstack) or side by side (hstack)

- squeeze -> removes all 1D from a tensor
- unsqueeze -> add 1D to a target tensor

- permute -> return a view of the input with dimensions permuted (swapped) in a certain way


````python
import torch

x = torch.arange(1.,10.)
x, x.shape
````

##### Reshape

````python
#add an extra dimension

xr = x.reshape(1,9)
xr, xr.shape, xr.ndim
````

##### View

````python
z = x.view(1,9)
z.shap
````

````python
z[:,0] = 5
z, x
````

##### Stacking

- hstack uses dim = 0,
- vstack uses dim = 1

````python
xs = torch.stack([x,x,x,x], dim = 1)
xs
````

##### Permute

- Used widely with images

````python
y = torch.rand(2,3,4)
y.size()
````

````python
torch.permute(y,(2,0,1)).size()
````

````python
xorig = torch.rand(size = (224, 224, 3))

xperm = xorig.permute(2,0,1)  #swaps the position of the axis

print(f"Previous shape: {xorig.shape}")
print(f"New shape: {xperm.shape}")
````

### Indexing (selecting data from Tensors)

Indexing with Pytorch is similar to indexing with Numpy.

````python
import torch
x = torch.arange(1,10).reshape(1,3,3)
x, x.shape
````

````python
#index on the middle bracket
x[0][0]
````

````python
#lets index on the most inner bracket (last dim)
x[0][1][1]
````

````python
# : can also be used to select either row or col
x[:,0]
````

````python
#get all values of 0th and 1st dim but only index 1 of 2nd dimension

x[:,:,1]
````

````python
#get all values of 0th and 1st dim but only 1 index of 1st and 2nd dim
x[:,1,1]
````

````python
#get index 0 of the 0th and 1st dim and all values of 2nd dim
x[0,0,:]
````

````python
#print 3,6,9

x[:,:,2]
````
